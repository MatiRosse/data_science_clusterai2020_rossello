# IMPORTAR LIBRERÍAS

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Se agregó la codificación utf-8 y latin para que lea el código correctamente
dengue_18_19=pd.read_csv("vigilancia-de-dengue-y-zika_2019.csv", encoding = 'UTF-8')
dengue_20=pd.read_csv("vigilancia-de-dengue-y-zika_2020.csv", encoding = 'latin')
observatorios=pd.read_csv("Exp_183015_observatorios.csv", encoding = 'latin')
precipitaciones=pd.read_csv("Exp_183015_precipitaciones.csv", encoding = 'latin')
clima=pd.read_csv("Exp_183015_horarios.csv", encoding = 'latin')
dengue_18_19.head(20)
```

# LIMPIEZA CLIMA

```python
#Convertimos la columna FECHA en formato fecha, y agregamos una columna "Mes" y "Año"
clima.FECHA = pd.to_datetime(clima.FECHA, dayfirst=True ) 
clima['MES'] = clima.FECHA.dt.month 
clima['AÑO'] = clima.FECHA.dt.year 
clima.head()
```

```python
#Eliminamos la columna FECHA y HORA LOCAL ya que sólo filtramos por mes y año
clima  = clima.drop(['FECHA','HORA LOCAL'], axis=1)
clima
```

```python
#Obtenido ya el mes y año, agregamos los datos de precipitaciones de cada observatorio
clima_p = pd.merge(clima, precipitaciones, left_on=["ESTACION","MES","AÑO"], right_on=["ESTACION","MES","AÑO"], how="left")
clima_p
```

```python
clima_p.rename(columns={'CANTIDAD (mm)':'PRECIPITACION (mm)'}, inplace=True)

#Eliminamos la columna CODIGO PRECIP que no es de interés
clima_p= clima_p.drop('CODIGO PRECIP', axis=1)
clima_p
```

```python
#Verificamos si existen registros nulos
col_NaN1 = clima.isnull().any()
col_NaN1
```

```python
#Contamos cuantos registros nulos hay por variable
col_NaN1 = clima.isnull().sum()
col_NaN1
```

```python
#Eliminamos los registros con valores nulos
clima.dropna(how ='any', inplace = True)
np.shape(clima)
print(f'Nuestro dataset luego de la limpieza tiene {clima.shape[0]} filas y {clima.shape[1]} columnas.')
```

```python
#Agregamos a qué provincia pertenece cada observatorio con el listado de observatorios
clima_obs = pd.merge(clima_p, observatorios, left_on="ESTACION", right_on="ESTACION", how="inner")
clima_obs
```

```python
#Verificamos si todavía tenemos las 25 provincias.
np.unique(clima_obs.PROVINCIA)
np.unique(clima_obs.PROVINCIA).shape
print(f'En nuestro dataset luego de limpiar los nulos hay {np.unique(clima_obs.PROVINCIA).shape[0]} provincias distintas')
```

```python
clima_obs=clima_obs.loc[:, ['PROVINCIA', 'MES','AÑO','TEMPERATURA (ºC)','PRECIPITACION (mm)','HR (%)','INTENSIDAD DE VIENTO (km/h)']]
clima_obs
```

```python
#Buscamos las condiciones medias por mes de cada provincia en cada año
clima_obs=clima_obs.groupby(['PROVINCIA','MES','AÑO'], as_index=True ).mean()
clima_obs
```

```python
clima_obs=clima_obs.reset_index()
```

```python
clima_obs
```

# LIMPIEZA DENGUE DATAFRAME

```python
dengue_18_19.head(20)
```

```python
dengue_20.head()
```

```python
#Observamos que ambas tienen las mismas columnas, pero tenemos que renombrar anio para poder unirlas
dengue_20.rename(columns={'anio':'año'}, inplace=True)
dengue_20
```

```python
dengue = pd.concat([dengue_18_19,dengue_20], axis=0)
dengue
```

```python
np.shape(dengue)
print(f'Nuestro dataset tiene {dengue.shape[0]} registros y {dengue.shape[1]} columnas.')
```

```python
dengue['departamento_nombre'].replace('*sin dato*', np.nan, inplace = True )
dengue.head()
```

```python
#Verificamos si existen registros nulos. Vemos que sólo existen en departamento y cantidad de casos.
#No se eliminarán debido no nos interesa el departamento, y los nulos en cantidad de casos se reemplazarán por 0.
col_NaN = dengue.isnull().any()
col_NaN
```

```python
#Eliminamos en caso de que hayan duplicados
dengue.drop_duplicates()
```

```python
#Vamos a crear un brand de meses según la semana epidemológica para ubicarlo en el mes.
dengue.loc[(dengue['semanas_epidemiologicas'] > 0) & (dengue['semanas_epidemiologicas'] <= 5), 'Mes'] = 1
dengue.loc[(dengue['semanas_epidemiologicas'] > 5) & (dengue['semanas_epidemiologicas'] <= 9 ), 'Mes'] = 2
dengue.loc[(dengue['semanas_epidemiologicas'] > 9) & (dengue['semanas_epidemiologicas'] <= 14), 'Mes'] = 3
dengue.loc[(dengue['semanas_epidemiologicas'] > 14) & (dengue['semanas_epidemiologicas'] <= 18), 'Mes'] = 4
dengue.loc[(dengue['semanas_epidemiologicas'] > 18) & (dengue['semanas_epidemiologicas'] <= 23), 'Mes'] = 5
dengue.loc[(dengue['semanas_epidemiologicas'] > 23) & (dengue['semanas_epidemiologicas'] <= 27), 'Mes'] = 6
dengue.loc[(dengue['semanas_epidemiologicas'] > 27) & (dengue['semanas_epidemiologicas'] <= 31), 'Mes'] = 7
dengue.loc[(dengue['semanas_epidemiologicas'] > 31) & (dengue['semanas_epidemiologicas'] <= 36), 'Mes'] = 8
dengue.loc[(dengue['semanas_epidemiologicas'] > 36) & (dengue['semanas_epidemiologicas'] <= 40), 'Mes'] = 9
dengue.loc[(dengue['semanas_epidemiologicas'] > 40) & (dengue['semanas_epidemiologicas'] <= 44), 'Mes'] = 10
dengue.loc[(dengue['semanas_epidemiologicas'] > 44) & (dengue['semanas_epidemiologicas'] <= 49), 'Mes'] = 11
dengue.loc[(dengue['semanas_epidemiologicas'] > 50) & (dengue['semanas_epidemiologicas'] <= 53), 'Mes'] = 12

dengue=dengue.sort_values(['semanas_epidemiologicas'], ascending=[True])
dengue.head()
```

```python
dengue=dengue.drop(['departamento_nombre','departamento_id','provincia_id','semanas_epidemiologicas','evento_nombre','grupo_edad_id','grupo_edad_desc'], axis=1)
dengue.head(20)
```

```python
#Realizamos una agrupación para poder obtener la cantidad de casos totales en cada provincia en cada mes
dengue=dengue.groupby(['provincia_nombre','Mes','año'], as_index=True ).sum()
dengue.head(50)
```

```python
dengue=dengue.reset_index()
```

# UNION DENGUE Y CLIMA

```python
#Observamos la diferencia entre minúsculas y mayúsculas entre ambos dataframes a concatenar
#Por lo tanto vamos a modificar el formato del dengue, pasandolo todo a mayúscula
dengue.tail(20)
```

```python
dengue["provincia_nombre"] = dengue["provincia_nombre"].apply(lambda name : name.upper())
dengue
```

```python
casos = pd.merge(clima_obs, dengue, left_on=['PROVINCIA','MES','AÑO'], right_on=['provincia_nombre','Mes','año'], how="left")
casos.head(20)
```

```python
casos.drop(columns=["provincia_nombre","Mes","año"], inplace=True)
casos.head(20)
```

```python
#Una vez que logrado hacer la unión, reemplazamos los NaN por 0, que serían las provincias que no tuvieron casos en un mes determinado.
casos.fillna(0, inplace=True)
casos.head(50)
```

```python
casos.cantidad_casos=casos.cantidad_casos.astype(int)
```

```python
casos.head(50)
```

```python
np.shape(casos)
print(f'Nuestro dataset luego de la limpieza tiene {casos.shape[0]} filas y {clima.shape[1]} columnas.')
```

```python
#Modificamos finalmente los nombres de nuestras columnas para trabajar luego en el EDA
casos.rename(columns={'PROVINCIA':'Provincia'}, inplace=True)
casos.rename(columns={'MES':'Mes'}, inplace=True)
casos.rename(columns={'AÑO':'Anio'}, inplace=True)
casos.rename(columns={'TEMPERATURA (ºC)':'Temperatura'}, inplace=True)
casos.rename(columns={'PRECIPITACION (mm)':'Precipitacion'}, inplace=True)
casos.rename(columns={'HR (%)':'Humedad'}, inplace=True)
casos.rename(columns={'INTENSIDAD DE VIENTO (km/h)':'Vel_vientos'}, inplace=True)
casos
```
